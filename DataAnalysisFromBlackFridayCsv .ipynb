```python
import pandas as pd
```

```python
df = pd.read_csv("BlackFriday.csv")
df
```

```python
#Analayzing Gender
#how many rows are there for male and female
len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])
```

```python
#Representing it i graph
data = pd.DataFrame({'Ratio' :[len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]} , index = ['Male','Female'])
```

```python
data.plot.pie(y = 'Ratio')
```

```python
#using groupby function 
df.groupby('Gender').size().plot(kind = 'bar', figsize = (6,6))
```

```python
#finding how much sum of money did each gender spent
df.groupby('Gender').sum()['Purchase']
```

```python
#finding the average of how much they spend
df.groupby('Gender').mean()['Purchase'].plot(kind = 'pie', autopct = "%0.1f")
```
