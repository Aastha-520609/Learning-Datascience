```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
#used for static plots
%matplotlib inline 
```

```python
#1-Generating cos and sin graph
x = np.linspace(0, 4*np.pi, 100)
sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')
plt.show()

cos_y = np.cos(x)
plt.plot(x,cos_y)
plt.title('Cos Wave')
plt.show()

sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()

#giving graph line some different pattern
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , '*', label = 'Sine')
plt.plot(x,cos_y, '--', label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()
```

```python
plt.style.use('grayscale')

sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()
```

```python
#Increasing the size of the plot
plt.figure(figsize = (18,5))

plt.style.use('grayscale')

sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()
```

```python
#3-Saving graph in our local system
plt.figure(figsize = (18,5))

plt.style.use('seaborn-dark')

fig = plt.figure()

plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')

plt.title('Combined Sine and Cos Wave')

#saving
fig.savefig('Sine & Cosine.png')
```

```python
#D- linestyle command
plt.figure(figsize = (15,5))

plt.plot(x, x+0, linestyle = 'solid')
plt.plot(x, x+1, linestyle = 'dashed')
plt.plot(x, x+2, linestyle = 'dashdot')

plt.show()
```

```python
#E- Adusting plot-changing the dimension

x = np.linspace(0, 4*np.pi, 100)

plt.xlim([0,15])
plt.ylim([-2,2])

sin_y = np.sin(x)
plt.plot(x, sin_y)

plt.title('Sine Wave')
plt.show()
```

```python
#Using set function
x = np.linspace(0, 4*np.pi, 100)
fig = plt.axes()
fig.plot(x, np.sin(x))
fig.set(xlim = (0,10), ylim = (-2.5, 2.5), xlabel = 'x', ylabel = 'sin(x)', title = 'Sie Wave')
fig.plot()
```

```python
#Scatter plot
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y, c = 'green', alpha = .6,  s = 10)
plt.title('Scatter plot')

plt.show()
```

```python
#multiple line plot
#on real datasets
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target

plt.scatter(x[:,2], x[:,3], c='blue', s=50, alpha = .9)

plt.xlabel(iris['feature_names'][2])
plt.ylabel(iris['feature_names'][3])
```

```python
#Histograms and binnings
#histograms displays the distribution of a set of continuous or discrete data

data = np.random.randn(1000)
plt.hist(data , bins = 30, edgecolor = 'white', alpha = .8, histtype = 'stepfilled') # histtype is histogram type that you want to specify

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
#Multiple  histogram 
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)+1
x3 = np.random.randn(1000)+2

plt.hist(x1, bins = 20, alpha = .8)
plt.hist(x2, bins = 20, alpha = .8)
plt.hist(x3, bins = 20, alpha = .8)

plt.title('Random Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
## 2D histograms - provides top view of an histogram
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)

plt.hist2d(x1,x2,  bins = 20)

plt.title('2d Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
#PIE charts
cars = ['AUDI', 'BMW', 'FORD', 'TESLA']
data = [400, 300, 1000, 10]

fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars)
plt.show()
```

```python
#Basic operations on pie chart
cars = ['AUDI', 'BMW', 'FORD', 'TESLA']
data = [400, 300, 1000, 10]
explode = [0, 0, 0, .1] #is used to make some element out of piechart

fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars, explode = explode)
plt.show()
```

```python
#adding shadow , percentile
cars = ['AUDI', 'BMW', 'FORD', 'TESLA']
data = [400, 300, 1000, 10]
explode = [0, 0, 0, .1] #is used to make some element out of piechart

fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars, explode = explode,)
plt.show()
```
