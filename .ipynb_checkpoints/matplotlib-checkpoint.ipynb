```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
#used for static plots
%matplotlib inline 
```

```python
#1-Generating cos and sin graph
x = np.linspace(0, 4*np.pi, 100)
sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')
plt.show()

cos_y = np.cos(x)
plt.plot(x,cos_y)
plt.title('Cos Wave')
plt.show()

sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()

#giving graph line some different pattern
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , '*', label = 'Sine')
plt.plot(x,cos_y, '--', label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()
```

```python
#2- Setting styles to the graph
plt.style.available
```

```python
plt.style.use('grayscale')

sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()
```

```python
#Increasing the size of the plot
plt.figure(figsize = (18,5))

plt.style.use('grayscale')

sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')
plt.title('Combined Sine and Cos Wave')
plt.legend() # shows colour index
plt.show()
```

```python
#3-Saving graph in our local system
plt.figure(figsize = (18,5))

plt.style.use('seaborn-dark')

fig = plt.figure()

plt.plot(x, sin_y , label = 'Sine')
plt.plot(x,cos_y, label = 'Cosine')

plt.title('Combined Sine and Cos Wave')

#saving
fig.savefig('Sine & Cosine.png')
```

```python
#D- linestyle command
plt.figure(figsize = (15,5))

plt.plot(x, x+0, linestyle = 'solid')
plt.plot(x, x+1, linestyle = 'dashed')
plt.plot(x, x+2, linestyle = 'dashdot')

plt.show()
```

```python
#E- Adusting plot-changing the dimension

x = np.linspace(0, 4*np.pi, 100)

plt.xlim([0,1])
plt.ylim([-2,2])

sin_y = np.sin(x)
plt.plot(x, sin_y)

plt.title('Sine Wave')
plt.show()
```
