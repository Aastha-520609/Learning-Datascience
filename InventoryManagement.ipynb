```python
import time

#opening the inventory txt file in read mode
fd = open("Inventory.txt", "r")
txt = fd.read().split("\n")
fd.close()

#Taking user input
user_name = input("Enter your Name: ")
phone_number = input("Enter Your Phone Number: ")
prod_ID = input("Enter the product id: ")
prod_quan = input("Enter the product Quantity:")

#Creating an empty list
updatedProductList = []

#Going through each line in txt file
for product in txt:
    prod_details = product.split(",")
    
    #checking if product exists or not
    if(prod_details[0] == prod_ID):
        if(int(prod_quan) <= int(prod_details[3])):
            
            print("Product Name: ", prod_details[1])
            print("Price: ", prod_details[2])
            print("Quantity: ", prod_quan)
            print("Total Billing Amount: ", int(prod_quan) * int(prod_details[2]))
        
            #updating product quantity
            prod_details[3] = str(int(prod_details[3]) - int(prod_quan))
        
            #Adding transaction into sales.txt file
            fd = open("salesFile.txt", 'a')
            sales_details = user_name + "," + phone_number + "," + prod_details[1] + "," + prod_ID + "," + prod_quan + "," + str(int(prod_quan) * int(prod_details[2])) + "," + time.ctime() + "," + "\n" 
            fd.write(sales_details)
            fd.close()

    else:
        #If we're not having enough quantity
        
        print("Sorry, We're not having enough quantity.")
        print("We are only having", prod_details[3], 'quantity.')
        print("Would you like to purchase it?")
        
        ch = input("Press Y/N?")
        
        if(ch == 'Y' or ch == 'y'):
            
            print("Product Name: ", prod_details[1])
            print("Price       : ", prod_details[2])
            print("Quantity    : ", prod_details[3])
            print("Billing Amount :", int(prod_details[3]) * int(prod_details[2]))
            
            
            prod_details[3] = '0'
        
        else:
            print("Thanks")
    
    #updating whole line of particular product id that has been bought
    updatedProductList.append(prod_details)

#updating and writing back into the inventory txt file    
lst = []
for i in updatedProductList:
     prod = i[0] +","+  i[1] +","+ i[2] +","+ i[3] + '\n'
     lst.append(prod)

lst[-1] = lst[-1][:-1] #removing last \n 

fd = open("Inventory.txt", "w")

for i in lst:
    fd.write(i)

fd.close()

print("Updated")
```
