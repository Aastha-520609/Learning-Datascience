```python
import numpy as np 
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df
```

```python
#1st type of random sampling - simple random sampling
#showing random sample using sample function
random_sample = df.sample(10)
random_sample
```

```python
#2nd type - systematic sampling - members are chosen at regular intervals
df.iloc[np.arange(0,len(df),20)]
```

```python
#3rd sampling - stratified sampling
#disproportionate startified samplling
df.groupby("origin", group_keys = False).apply(lambda x: x.sample(5))
```

```python
#to get sample according to the population
#proportionate startified samplling
df.groupby("origin", group_keys = False).apply(lambda x: x.sample(frac = 0.1))
```

# Central Limit Theorem

```python
sns.displot(df["displacement"], kind = "kde")
```

```python
#taking sample values and finding the displacement peak
sample_mean_list = []
for i in(0,30):
    random_sample = df["displacement"].sample(30)
    sample_mean_list.append(random_sample.mean())
```

```python
sns.displot(sample_mean_list, kind = "kde")
```
