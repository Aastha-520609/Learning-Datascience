# Handling null values

```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
#Usinng imputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
data['Rating'].head(50)
```

```python
imputer.fit(data.iloc[ : , 2:3 ].values)
data.iloc[ : , 2:3 ] = imputer.transform(data.iloc[ : , 2:3 ].values)

df = data.dropna()
df
```

# Data Analysis using GroupBy

```python
#Naming the apps that are art and design with rating more than 4.5 in descending order wrt rating
process = df[df['Category'] == 'ART_AND_DESIGN']
process = process[process['Rating'] > 4.5]

process.sort_values(by = 'Rating', ascending = True).head()
```
